```python
import json

file_path = "/workspace/processed_data/0_10000.jsonl"

total_words = 0
num_docs = 0

with open(file_path, "r", encoding="utf-8") as f:
    for line in f:
        data = json.loads(line)
        content = data.get("content", "")
        if not content:
            continue
        word_count = len(content.split())  # split on whitespace
        total_words += word_count
        num_docs += 1

average_words = total_words / num_docs if num_docs > 0 else 0
print(f"Average words per document: {average_words:.2f}")
```
