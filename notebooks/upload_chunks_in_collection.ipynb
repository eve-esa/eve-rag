```python
import os
import glob
import subprocess
import sys

folder_path = "/workspace/processed_data2"
jsonl_files = sorted(glob.glob(os.path.join(folder_path, "*.jsonl")))

project_root = "/workspace/eve-rag"

env = os.environ.copy()
env["PYTHONPATH"] = project_root  # ensure Python can find 'data'

for file_path in jsonl_files[:4]:
    print(f"Processing {file_path}")
    subprocess.run(
        [sys.executable, "-m", "data.upload_chunks_qdrant", "--file-path", file_path],
        cwd=project_root,
        env=env
    )
    print(f"Finished {os.path.basename(file_path)}\n")
```
